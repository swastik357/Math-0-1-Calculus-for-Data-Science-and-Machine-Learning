```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.linspace(-2, 2, 100)
y = np.exp(x)

def exp_approx(x, N):
  y_all = np.zeros((N + 1, len(x)))
  for n in range(N + 1):
    y_all[n] = x**n / math.factorial(n)
  return y_all.sum(axis=0)
```

```python
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.title("Convergence of Taylor series to true function with increase in polynomial degree")
plt.plot(x, y, label='true function value',linewidth=3, color='lightblue', marker='o', markersize=6, markevery=10);
for n in range(1,6):
  y2 = exp_approx(x, n)
  plt.plot(x, y2, label='approximation with polynomial degree = {}'.format(n));
plt.legend();
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)

def sin_approx(x, N):
  y_all = np.zeros((N + 1, len(x)))
  for n in range(N + 1):
    coef = 1 if n % 2 == 0 else -1
    nth_term = coef * (x ** (2 * n + 1)) / math.factorial(2 * n + 1)
    y_all[n] = nth_term
  return y_all.sum(axis=0)
```

```python
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.plot(x, y, label='true function value',linewidth=3, color='lightblue', marker='o', markersize=6, markevery=10);
plt.title("Convergence of Taylor series to true function with increase in polynomial degree")
for n in range(1,4):
  y2 = sin_approx(x, n)
  plt.plot(x, y2, label='approximation with polynomial degree = {}'.format(n));
plt.legend();
```

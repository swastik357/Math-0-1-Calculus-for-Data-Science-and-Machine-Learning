```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 100)
y = 0.5 * x + 1
```

```python
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')
plt.grid()
plt.plot(x, y, label = '0.5 * x + 1');
plt.legend()
```

```python
x = np.linspace(-2, 2, 100)
y = -0.5 * 9.8 * x ** 2 + 2 * x + 1
```

```python
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')
plt.grid()
plt.plot(x, y, label = '-0.5 * 9.8 * x^2 + 2 * x + 1');
plt.legend()
```

```python
x = np.linspace(0, 10, 100)
y = -0.5 * x ** (1./3) + 0.2 * x ** (1./2) + 1
```

```python
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')
plt.grid()
plt.plot(x, y, label = '-0.5 * x^(1./3) + 0.2 * x^(1./2) + 1');
plt.legend()
```

```python
x = np.linspace(-5, 5, 100)
y = np.exp(x) # e ** x
```

```python
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')
plt.grid()
plt.plot(x, y, label = 'e^x');
plt.legend()
```

```python
x = np.linspace(-5, 5, 100)
y = np.exp(x) # e ** x
y2 = x**2
```

```python
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')
plt.grid()
plt.plot(x, y, label='e^x');
plt.plot(x, y2, label='x^2');
plt.legend();
```

```python
x = np.linspace(-5 * np.pi, 5 * np.pi, 200)
y = np.sin(x)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 200)
y = np.sin(x)
y2 = np.cos(x)
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.tan(x)
```

```python
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')
plt.grid()
plt.ylim(bottom=-10, top=10)
plt.plot(x, y, label = 'tan');
plt.legend()
```
